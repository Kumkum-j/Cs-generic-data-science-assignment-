```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv("International_T20_Data.csv")
# Set style
sns.set(style="whitegrid")




```

```python
# 1. histplot()
# Use: Shows frequency distribution of a numerical column
# Syntax: sns.histplot(data=df, x='column')
# ==========================================================
plt.figure(figsize=(6,4))
sns.histplot(data=df, x='info.outcome.by.runs', bins=20)
plt.title("Distribution of Runs")
plt.show()
```

```python
# 2. boxplot()
# Use: Detects spread and outliers in numerical data
# Syntax: sns.boxplot(data=df, x='column')
# ==========================================================
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='info.outcome.by.wickets')
plt.title("Boxplot of Wickets")
plt.show()
```

```python
# 3. scatterplot()
# Use: Shows relationship between two numerical variables
# Syntax: sns.scatterplot(data=df, x='col1', y='col2')
# ==========================================================
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='info.outcome.by.runs', y='info.outcome.by.wickets')
plt.title("Runs vs Wickets")
plt.show()
```

```python
# 4. lineplot()
# Use: Displays trend over ordered values
# Syntax: sns.lineplot(data=df, x='col1', y='col2')
# ==========================================================
temp = df[['info.match_type_number','info.outcome.by.runs']].dropna().sort_values('info.match_type_number')
plt.figure(figsize=(6,4))
sns.lineplot(data=temp, x='info.match_type_number', y='info.outcome.by.runs')
plt.title("Match Number vs Runs")
plt.show()

```

```python
# 5. barplot()
# Use: Compares average values across categories
# Syntax: sns.barplot(data=df, x='category', y='value')
# ==========================================================
temp = df[['info.gender','info.outcome.by.runs']].dropna()
plt.figure(figsize=(6,4))
sns.barplot(data=temp, x='info.gender', y='info.outcome.by.runs')
plt.title("Average Runs by Gender")
plt.show()
```

```python
# 6. countplot()
# Use: Counts occurrences of categories
# Syntax: sns.countplot(data=df, x='category')
# ==========================================================
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='info.toss.decision')
plt.title("Toss Decision Count")
plt.show()
```

```python
# 7. violinplot()
# Use: Shows distribution + density of data
# Syntax: sns.violinplot(data=df, x='category', y='value')
# ==========================================================
temp = df[['info.gender','info.outcome.by.runs']].dropna()
plt.figure(figsize=(6,4))
sns.violinplot(data=temp, x='info.gender', y='info.outcome.by.runs')
plt.title("Runs Distribution by Gender")
plt.show()
```

```python
# 8. stripplot()
# Use: Displays all individual observations
# Syntax: sns.stripplot(data=df, x='category', y='value')
# ==========================================================
plt.figure(figsize=(6,4))
sns.stripplot(data=temp, x='info.gender', y='info.outcome.by.runs')
plt.title("Stripplot of Runs by Gender")
plt.show()
```

```python
# 9. swarmplot()
# Use: Prevents overlapping points in categorical scatter
# Syntax: sns.swarmplot(data=df, x='category', y='value')
# ==========================================================
plt.figure(figsize=(6,4))
sns.swarmplot(data=temp.sample(200), x='info.gender', y='info.outcome.by.runs')
plt.title("Swarmplot of Runs")
plt.show()
```

```python
# 10. heatmap()
# Use: Shows correlation matrix visually
# Syntax: sns.heatmap(df.corr())
# ==========================================================
plt.figure(figsize=(6,4))
corr = df[['info.outcome.by.runs','info.outcome.by.wickets','info.overs','info.match_type_number']].corr()
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 11. pairplot()
# Use: Shows pairwise relationships among numeric columns
# Syntax: sns.pairplot(df[['col1','col2']])
# ==========================================================
sns.pairplot(df[['info.outcome.by.runs','info.outcome.by.wickets','info.overs']].dropna())
plt.show()
```

```python
# 12. kdeplot()
# Use: Shows smooth probability density curve
# Syntax: sns.kdeplot(data=df, x='column')
# ==========================================================
plt.figure(figsize=(6,4))
sns.kdeplot(data=df, x='info.outcome.by.runs')
plt.title("Density Plot of Runs")
plt.show()
```

```python
# 13. regplot()
# Use: Shows regression trend between two variables
# Syntax: sns.regplot(data=df, x='col1', y='col2')
# ==========================================================
temp = df[['info.outcome.by.runs','info.outcome.by.wickets']].dropna()
plt.figure(figsize=(6,4))
sns.regplot(data=temp, x='info.outcome.by.runs', y='info.outcome.by.wickets')
plt.title("Regression Plot")
plt.show()
```

```python
# 14. catplot()
# Use: General categorical plotting function
# Syntax: sns.catplot(data=df, x='category', y='value', kind='box')
# ==========================================================

temp_cat = df[['info.gender', 'info.outcome.by.runs']].dropna()

sns.catplot(
    data=temp_cat,
    x='info.gender',
    y='info.outcome.by.runs',
    kind='box'
)

plt.title("Catplot of Runs by Gender")
plt.show()
```

```python
# 15. displot()
# Use: Distribution plot with histogram/kde
# Syntax: sns.displot(data=df, x='column')
# ==========================================================
sns.displot(data=df, x='info.outcome.by.runs')
plt.show()
```
