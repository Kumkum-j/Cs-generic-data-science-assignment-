```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv("International_T20_Data.csv")
```

```python
# 1. plt.plot()
# Use: Draws line graph to show trend between two variables
# Syntax: plt.plot(x, y)
# ==========================================================
temp = df[['info.match_type_number','info.outcome.by.runs']].dropna().sort_values('info.match_type_number')

plt.figure(figsize=(6,4))
plt.plot(temp['info.match_type_number'], temp['info.outcome.by.runs'])
plt.title("Line Plot: Match Number vs Runs")
plt.xlabel("Match Number")
plt.ylabel("Runs")
plt.show()
```

```python
# 2. plt.scatter()
# Use: Shows relationship between two numerical columns
# Syntax: plt.scatter(x, y)
# ==========================================================
temp = df[['info.outcome.by.runs','info.outcome.by.wickets']].dropna()

plt.figure(figsize=(6,4))
plt.scatter(temp['info.outcome.by.runs'], temp['info.outcome.by.wickets'])
plt.title("Scatter Plot: Runs vs Wickets")
plt.xlabel("Runs")
plt.ylabel("Wickets")
plt.show()
```

```python
# 3. plt.bar()
# Use: Compares categorical values
# Syntax: plt.bar(x, height)
# ==========================================================
temp = df['info.gender'].value_counts()

plt.figure(figsize=(6,4))
plt.bar(temp.index, temp.values)
plt.title("Bar Plot: Gender Count")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
# 4. plt.hist()
# Use: Shows frequency distribution
# Syntax: plt.hist(data)
# ==========================================================
plt.figure(figsize=(6,4))
plt.hist(df['info.outcome.by.runs'].dropna(), bins=20)
plt.title("Histogram of Runs")
plt.xlabel("Runs")
plt.ylabel("Frequency")
plt.show()
```

```python
# 5. plt.boxplot()
# Use: Detects spread and outliers
# Syntax: plt.boxplot(data)
# ==========================================================
plt.figure(figsize=(6,4))
plt.boxplot(df['info.outcome.by.wickets'].dropna())
plt.title("Boxplot of Wickets")
plt.show()
```

```python
# 6. plt.pie()
# Use: Shows category proportion
# Syntax: plt.pie(values, labels=labels)
# ==========================================================
temp = df['info.toss.decision'].value_counts()

plt.figure(figsize=(6,4))
plt.pie(temp.values, labels=temp.index, autopct='%1.1f%%')
plt.title("Pie Chart of Toss Decision")
plt.show()
```

```python
# 7. plt.stem()
# Use: Displays discrete data points
# Syntax: plt.stem(x, y)
# ==========================================================
temp = df[['info.match_type_number','info.outcome.by.runs']].dropna().head(20)

plt.figure(figsize=(6,4))
plt.stem(temp['info.match_type_number'], temp['info.outcome.by.runs'])
plt.title("Stem Plot")
plt.show()
```

```python
# 8. plt.step()
# Use: Shows stepwise trend
# Syntax: plt.step(x, y)
# ==========================================================
plt.figure(figsize=(6,4))
plt.step(temp['info.match_type_number'], temp['info.outcome.by.runs'])
plt.title("Step Plot")
plt.show()
```

```python
# 9. plt.fill_between()
# Use: Fills area under line
# Syntax: plt.fill_between(x, y)
# ==========================================================
plt.figure(figsize=(6,4))
plt.fill_between(temp['info.match_type_number'], temp['info.outcome.by.runs'])
plt.title("Fill Between Plot")
plt.show()
```

```python
# 10. plt.errorbar()
# Use: Shows values with error bars
# Syntax: plt.errorbar(x, y, yerr=error)
# ==========================================================
plt.figure(figsize=(6,4))
plt.errorbar(temp['info.match_type_number'], temp['info.outcome.by.runs'], yerr=2)
plt.title("Error Bar Plot")
plt.show()
```

```python
# 11. plt.stackplot()
# Use: Shows cumulative stacked data
# Syntax: plt.stackplot(x, y1, y2)
# ==========================================================
temp = df[['info.match_type_number','info.outcome.by.runs','info.outcome.by.wickets']].dropna().head(20)

plt.figure(figsize=(6,4))
plt.stackplot(temp['info.match_type_number'],
              temp['info.outcome.by.runs'],
              temp['info.outcome.by.wickets'])
plt.title("Stack Plot")
plt.show()
```

```python
# 12. plt.vlines()
# Use: Draws vertical lines
# Syntax: plt.vlines(x, ymin, ymax)
# ==========================================================
plt.figure(figsize=(6,4))
plt.vlines(temp['info.match_type_number'], 0, temp['info.outcome.by.runs'])
plt.title("Vertical Lines Plot")
plt.show()
```

```python
# 13. plt.hlines()
# Use: Draws horizontal lines
# Syntax: plt.hlines(y, xmin, xmax)
# ==========================================================
plt.figure(figsize=(6,4))
plt.hlines(temp['info.outcome.by.runs'], 0, temp['info.match_type_number'])
plt.title("Horizontal Lines Plot")
plt.show()
```

```python
# 14. plt.hexbin()
# Use: Density plot for large scatter data
# Syntax: plt.hexbin(x, y)
# ==========================================================
temp = df[['info.outcome.by.runs','info.outcome.by.wickets']].dropna()

plt.figure(figsize=(6,4))
plt.hexbin(temp['info.outcome.by.runs'], temp['info.outcome.by.wickets'])
plt.title("Hexbin Plot")
plt.show()
```

```python
# 15. plt.eventplot()
# Use: Shows occurrence of events
# Syntax: plt.eventplot(data)
# ==========================================================
plt.figure(figsize=(6,4))
plt.eventplot(df['info.outcome.by.runs'].dropna().head(30))
plt.title("Event Plot")
plt.show()
```
