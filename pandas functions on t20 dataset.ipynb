```python
df = pd.read_csv("International_T20_Data.csv")
```

```python
# 1. head()
# Use: Shows first rows of dataset
# Syntax: df.head(n)
# ==========================================================
print("First 5 rows:")
print(df.head())
```

```python
# 2. tail()
# Use: Shows last rows of dataset
# Syntax: df.tail(n)
# ==========================================================
print("Last 5 rows:")
print(df.tail())
```

```python
# 3. info()
# Use: Gives column names, datatype, null values
# Syntax: df.info()
# ==========================================================
print("Dataset Info:")
df.info()
```

```python
# 4. describe()
# Use: Statistical summary of numeric columns
# Syntax: df.describe()
# ==========================================================
print("Statistical Summary:")
print(df.describe())
```

```python
# 5. isnull()
# Use: Detects missing values
# Syntax: df.isnull()
# ==========================================================
print("Missing Values Count:")
print(df.isnull().sum())
```

```python
# 6. dropna()
# Use: Removes missing values
# Syntax: df.dropna()
# ==========================================================
clean_df = df[['info.outcome.by.runs', 'info.outcome.by.wickets']].dropna()
print("After Dropna:")
print(clean_df.head())
```

```python
# 7. fillna()
# Use: Replaces missing values
# Syntax: df.fillna(value)
# ==========================================================
filled_runs = df['info.outcome.by.runs'].fillna(0)
print("Filled Missing Runs:")
print(filled_runs.head())
```

```python
# 8. sort_values()
# Use: Sorts data by column
# Syntax: df.sort_values('column')
# ==========================================================
sorted_runs = df.sort_values('info.outcome.by.runs')
print("Sorted Runs:")
print(sorted_runs[['info.outcome.by.runs']].head())
```

```python
# 9. value_counts()
# Use: Counts unique values in categorical column
# Syntax: df['column'].value_counts()
# ==========================================================
print("Gender Count:")
print(df['info.gender'].value_counts())
```

```python
# 10. groupby()
# Use: Groups data and performs aggregation
# Syntax: df.groupby('column').mean()
# ==========================================================
grouped = df.groupby('info.gender')['info.outcome.by.runs'].mean()
print("Average Runs by Gender:")
print(grouped)
```

```python
# 11. loc[]
# Use: Select rows/columns by label
# Syntax: df.loc[row, column]
# ==========================================================
print("Using loc:")
print(df.loc[0:5, ['info.gender', 'info.outcome.by.runs']])
```

```python
# 12. iloc[]
# Use: Select rows/columns by index position
# Syntax: df.iloc[row, column]
# ==========================================================
print("Using iloc:")
print(df.iloc[0:5, 0:3])
```

```python
# 13. rename()
# Use: Renames columns
# Syntax: df.rename(columns={'old':'new'})
# ==========================================================
renamed_df = df.rename(columns={'info.gender':'Gender'})
print("Renamed Column:")
print(renamed_df[['Gender']].head())
```

```python
# 14. drop()
# Use: Removes column or row
# Syntax: df.drop('column', axis=1)
# ==========================================================
dropped_df = df.drop('info.gender', axis=1)
print("After Dropping Column:")
print(dropped_df.head())
```

```python
# 15. corr()
# Use: Finds correlation between numeric columns
# Syntax: df.corr()
# ==========================================================
corr_data = df[['info.outcome.by.runs','info.outcome.by.wickets','info.overs']].corr()
print("Correlation Matrix:")
print(corr_data)
```
