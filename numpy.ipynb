```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
uploaded = files.upload()
```

```python
# Load dataset
df = pd.read_csv("International_T20_Data.csv")
# Numeric data column
runs = df['info.outcome.by.runs'].dropna().values
wickets = df['info.outcome.by.wickets'].dropna().values
overs = df['info.overs'].dropna().values

```

```python
# 1. np.array()
# Use: Converts data into NumPy array
# Syntax: np.array(data)
# ==========================================================
runs_array = np.array(runs)
print("Array:", runs_array[:10])
```

```python
# 2. np.mean()
# Use: Calculates average value
# Syntax: np.mean(array)
# ==========================================================
print("Mean Runs:", np.mean(runs))
```

```python
# 3. np.median()
# Use: Finds middle value
# Syntax: np.median(array)
# ==========================================================
print("Median Runs:", np.median(runs))
```

```python
# 4. np.std()
# Use: Calculates standard deviation
# Syntax: np.std(array)
# ==========================================================
print("Standard Deviation:", np.std(runs))
```

```python
# 5. np.var()
# Use: Calculates variance
# Syntax: np.var(array)
# ==========================================================
print("Variance:", np.var(runs))

```

```python
# 6. np.min()
# Use: Finds minimum value
# Syntax: np.min(array)
# ==========================================================
print("Minimum Runs:", np.min(runs))
```

```python
# 7. np.max()
# Use: Finds maximum value
# Syntax: np.max(array)
# ==========================================================
print("Maximum Runs:", np.max(runs))

```

```python
# 8. np.sum()
# Use: Calculates total sum
# Syntax: np.sum(array)
# ==========================================================
print("Total Runs:", np.sum(runs))
```

```python
# 9. np.sort()
# Use: Sorts data in ascending order
# Syntax: np.sort(array)
# ==========================================================
print("Sorted Runs:", np.sort(runs)[:10])
```

```python
# 10. np.unique()
# Use: Finds unique values
# Syntax: np.unique(array)
# ==========================================================
print("Unique Wickets:", np.unique(wickets))
```

```python
# 11. np.sqrt()
# Use: Calculates square root
# Syntax: np.sqrt(array)
# ==========================================================
print("Square Root of Runs:", np.sqrt(runs[:10]))
```

```python
# 12. np.log()
# Use: Calculates natural logarithm
# Syntax: np.log(array)
# ==========================================================
positive_runs = runs[runs > 0]
print("Log of Runs:", np.log(positive_runs[:10]))
```

```python
# 13. np.corrcoef()
# Use: Finds correlation between two arrays
# Syntax: np.corrcoef(array1, array2)
# ==========================================================
min_len = min(len(runs), len(wickets))
print("Correlation:\n", np.corrcoef(runs[:min_len], wickets[:min_len]))

```

```python
# 14. np.percentile()
# Use: Finds percentile value
# Syntax: np.percentile(array, percentile)
# ==========================================================
print("75th Percentile Runs:", np.percentile(runs, 75))
```

```python
# 15. np.where()
# Use: Conditional filtering
# Syntax: np.where(condition)
# ==========================================================
print("Runs > 50:", np.where(runs > 50))
```
